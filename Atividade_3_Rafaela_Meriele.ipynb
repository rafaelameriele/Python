```python
import pandas as pd
url = "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"
dados = pd.read_csv(url) 
dados
```

```python
# 1 - Metragem média por bairro

dados.groupby('Bairro').mean()["Metragem"]
```

```python
# 2 - Duas formas de relacionar dados por bairro

dados.groupby('Bairro').median()
```

```python
dados.sort_values(by='Bairro')
```

```python
# 3 - Fazer gráficos e aplicar nas colunas DF

import matplotlib.pyplot as plt

grafico_bairro = dados["Bairro"].value_counts()
grafico_metragem = dados["Metragem"].value_counts()
grafico_valor = dados["Valor"].value_counts()


plt.figure(figsize = ((22, 18)))

plt.subplot(3, 2, 1)
grafico_bairro.head(20).plot.bar()
plt.title("Primeiros 20 bairros", fontsize = 15)

plt.subplot(3, 2, 2)
grafico_bairro.tail(20).plot.bar()
plt.title("Últimos 20 bairros", fontsize = 15)

plt.subplot(3, 2, 3)
grafico_bairro.plot.hist()
plt.title("Bairro - Histograma", fontsize = 15)

plt.subplot(3, 2, 4)
grafico_metragem.plot.bar()
plt.title("Metragem", fontsize = 15)

plt.subplot(3, 2, 5)
grafico_valor.plot.line()
plt.title("Linhas de Valor", fontsize = 15)

plt.subplots_adjust(left=0.2, bottom=0.1, right=1, top=1, wspace=0.3, hspace=0.8)

plt.show()
```
